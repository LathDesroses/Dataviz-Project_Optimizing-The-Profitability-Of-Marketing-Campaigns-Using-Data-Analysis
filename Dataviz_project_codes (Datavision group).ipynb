# Data Cleaning and Transformation

```python
# Check for duplicate rows
duplicate_rows = df.duplicated().sum()

print(f"Number of duplicate rows: {duplicate_rows}")
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df = df.dropna(subset=["ID"])
df.isnull().sum()
```

```python
# Select numerical columns
numerical_cols = ["Income", "Recency", "MntWines", "MntFruits", "MntMeatProducts",
                  "MntFishProducts", "MntSweetProducts", "MntGoldProds",
                  "NumDealsPurchases", "NumWebPurchases", "NumCatalogPurchases",
                  "NumStorePurchases", "NumWebVisitsMonth"]

# Determine the number of rows needed for subplots
num_features = len(numerical_cols)
rows = (num_features // 3) + (num_features % 3 > 0)  # Adjust row count based on number of features

# Create subplots dynamically
fig, axes = plt.subplots(nrows=rows, ncols=3, figsize=(15, rows * 4))
axes = axes.flatten()

# Plot histograms with KDE (Kernel Density Estimate) to analyze distributions
for i, col in enumerate(numerical_cols):
    sns.histplot(df[col].dropna(), bins=30, kde=True, ax=axes[i])
    axes[i].set_title(f"Distribution of {col}")
    axes[i].set_xlabel(col)

# Remove unused axes in the grid (if any)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

## Observations:

- Right-skewed distribution:
        Most income values are concentrated below 100,000.
        A few extreme values (outliers) exist above 200,000, stretching the distribution.

- Implication for Imputation:
        Since the distribution is skewed, mean imputation is not ideal because it would be influenced by the high-income outliers.
        Instead, median imputation is the best choice as it is more robust against skewness and outliers.

```python
#Median imputation
df["Income"] = df["Income"].fillna(df["Income"].median())

# Check for missing values again
missing_values_after = df.isnull().sum()

# Display missing values summary
print("Missing values after imputation:\n", missing_values_after)
```

```python
# Define categorical and numerical variables
categorical_vars = ["Education", "Marital_Status", "Complain", "Response",
                    "AcceptedCmp1", "AcceptedCmp2", "AcceptedCmp3",
                    "AcceptedCmp4", "AcceptedCmp5"]

numerical_vars = ["Income", "Recency", "MntWines", "MntFruits", "MntMeatProducts",
                  "MntFishProducts", "MntSweetProducts", "MntGoldProds",
                  "Kidhome", "Teenhome", "NumDealsPurchases", "NumWebPurchases",
                  "NumCatalogPurchases", "NumStorePurchases", "NumWebVisitsMonth"]

# Descriptive statistics with advanced analysis
display("Descriptive Statistics:\n", df[numerical_vars].describe())
```

```python
# Skewness & Kurtosis
skewness = df[numerical_vars].skew()
kurtosis = df[numerical_vars].kurt()
print("Skewness:\n", skewness)
print("Kurtosis:\n", kurtosis)
```

```python
# Interpretation of Skewness and Kurtosis with Recommendations
skew_kurt_table = pd.DataFrame({
    'Variable': numerical_vars,
    'Skewness': skewness.values,
    'Kurtosis': kurtosis.values,
    'Transformation Recommendation': [
        'Log Transform' if sk > 1 else 'Standardization' for sk in skewness
    ]
})
print("Skewness & Kurtosis Interpretation:\n", skew_kurt_table)

# Explanation of Transformations
print("\nExplanation of Transformations:")
print("- Log Transform: Applied to variables with high skewness (>1) to reduce extreme values and normalize the distribution.")
print("- Standardization: Applied to normally distributed or slightly skewed variables to ensure uniform scaling for clustering models like SOM.")
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

# Apply Log Transformation for all skewed variables (high and moderate skewness)
log_transform_vars = [
    "Income", "MntFruits", "MntMeatProducts", "MntSweetProducts", "NumDealsPurchases",
    "MntWines", "MntFishProducts", "MntGoldProds", "NumWebPurchases", "NumCatalogPurchases"
]
for col in log_transform_vars:
    df[col + "_log"] = np.log1p(df[col])  # Log(1+x) to handle 0 values safely

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from minisom import MiniSom
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Select variables for segmentation (no need for log transform)
segmentation_vars = [
    "Income", "MntFruits", "MntMeatProducts", "MntSweetProducts",
    "NumDealsPurchases", "MntWines", "MntFishProducts", "MntGoldProds",
    "NumWebPurchases", "NumCatalogPurchases", "Recency", "Kidhome",
    "Teenhome", "NumStorePurchases", "NumWebVisitsMonth", "Total_Spending", "Age", "Kids"
]

df_segmentation = df[segmentation_vars].copy()

# Standardize data
scaler = StandardScaler()
df_segmentation_scaled = scaler.fit_transform(df_segmentation)
```

## Train the Self-Organizing Map (SOM)

```python
# Define SOM grid size (5x5)
som_shape = (5, 5)

# Initialize SOM
som = MiniSom(som_shape[0], som_shape[1], df_segmentation_scaled.shape[1], sigma=1.0, learning_rate=0.5)

# Random weight initialization & training
som.random_weights_init(df_segmentation_scaled)
som.train_random(df_segmentation_scaled, num_iteration=1000)

# Assign each customer to a Best Matching Unit (BMU)
customer_segments = np.array([som.winner(x) for x in df_segmentation_scaled])

```

```python
def compute_u_matrix(som, data):
    """Compute the U-Matrix (distance map) for SOM visualization."""
    weights = som._weights
    m, n, dim = weights.shape
    u_matrix = np.zeros((m, n))

    for i in range(m):
        for j in range(n):
            neighbors = []
            if i > 0:
                neighbors.append(weights[i-1, j])
            if i < m-1:
                neighbors.append(weights[i+1, j])
            if j > 0:
                neighbors.append(weights[i, j-1])
            if j < n-1:
                neighbors.append(weights[i, j+1])

            u_matrix[i, j] = np.mean([np.linalg.norm(weights[i, j] - neighbor) for neighbor in neighbors])

    return u_matrix

# Compute U-Matrix
u_matrix = compute_u_matrix(som, df_segmentation_scaled)

# Visualize U-Matrix
plt.figure(figsize=(6, 6), facecolor="#F9FAF7")
plt.imshow(u_matrix, cmap="Blues", interpolation="nearest")
plt.colorbar()
plt.title("SOM U-Matrix (Cluster Separation)", fontsize=14, fontweight="bold", color="#3f456b")
plt.show()
```

## Apply K-Means on SOM Weights

#### Determine the Optimal Number of Clusters with K-Means

```python
# Reshape SOM weights for clustering
som_weights = som._weights.reshape(-1, df_segmentation_scaled.shape[1])

# Find optimal K using Elbow & Silhouette Method
k_values = range(2, 10)
inertia_values = []
silhouette_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10).fit(som_weights)
    inertia_values.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(som_weights, kmeans.labels_))

# Plot Elbow Method
plt.figure(figsize=(8, 4), facecolor="#F9FAF7")
plt.plot(k_values, inertia_values, marker='o', color="#3f456b")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia (Distortion)")
plt.title("Elbow Method for Optimal k", fontsize=14, fontweight="bold", color="#3f456b")
plt.show()

# Plot Silhouette Score
plt.figure(figsize=(8, 4), facecolor="#F9FAF7")
plt.plot(k_values, silhouette_scores, marker='o', color="#7580c6")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score for Optimal k", fontsize=14, fontweight="bold", color="#3f456b")
plt.show()

# Optimal clusters
n_clusters = 4  # Adjust based on elbow/silhouette
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10).fit(som_weights)

# Assign cluster labels
som_cluster_labels = kmeans.labels_.reshape(som_shape)

# Assign each customer to a segment
df.loc[:, "Segment"] = [som_cluster_labels[x, y] for x, y in customer_segments]
```

Based on your Elbow Method and Silhouette Score plots:
🔹 Elbow Method Analysis
The "elbow" point appears around k = 3 or k = 4, meaning the gain in inertia reduction slows down significantly after this.
Beyond k = 4, the curve flattens, indicating diminishing returns in reducing within-cluster variance.

🔹 Silhouette Score Analysis
The highest Silhouette Score is at k = 2 or k = 3, and it decreases significantly as k increases.
k = 3 has a good balance—beyond that, the score drops quickly, indicating that higher k values lead to less clearly separated clusters.

🔹 We choose → k = 4
